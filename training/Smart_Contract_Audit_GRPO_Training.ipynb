## Environment Setup

The training uses an OpenEnv-compatible Gymnasium environment (`SmartContractAuditEnv`) hosted on Hugging Face Spaces.

**Space:** https://huggingface.co/spaces/Gopichand0516/smart-contract-audit-env

The environment provides:
- **Observation space:** Solidity contract code (text)
- **Action space:** Vulnerability audit report (text)
- **Reward function:** Multi-component scoring (vulnerability detection + format compliance + severity accuracy)
